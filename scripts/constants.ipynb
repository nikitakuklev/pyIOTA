```python
import inspect
def recursive_print(el, prefix = None):
    prefix = prefix or []
    data = {}
    for attribute in el.__dict__:
        if attribute.isupper():
            val = getattr(el, attribute)    
            local_data = {} 
            if isinstance(val, (list, str)) and attribute != 'ALL':
                print(f'{">>".join(prefix+[attribute])} = {val}')
                data[attribute] = val
            elif inspect.isclass(val):
                print(val)
                local_data = recursive_print(val, prefix=prefix+[attribute])
                data[attribute] = local_data
    return data
data = recursive_print(RING)
        
```
